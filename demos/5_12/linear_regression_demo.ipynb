```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Linear regression with NumPy pseudoinverse

# We want: y ≈ X @ b
# b contains the intercept and slope.
# The first column of 1s lets us estimate the intercept.
X = np.array([
    [1, 4],
    [1, 8],
    [1, 7],
    [1, 5],
])
y = np.array([[4], [4], [5], [3]])

Xplus = np.linalg.pinv(X)
b = Xplus @ y

print("X =\n", X)
print("y =\n", y)
print("pseudoinverse of X =\n", Xplus)
print("b = Xplus @ y =\n", b)
print("intercept =", b[0, 0])
print("slope =", b[1, 0])

y_hat = X @ b
residuals = y - y_hat
sse = np.sum(residuals ** 2)
print("predictions y_hat =\n", y_hat)
print("residuals y - y_hat =\n", residuals)
print("SSE =", sse)

# Same answer using lstsq.
b_lstsq, residual_sum, rank, singular_values = np.linalg.lstsq(X, y, rcond=None)
print("\nUsing np.linalg.lstsq:")
print("b_lstsq =\n", b_lstsq)
print("residual_sum =", residual_sum)
print("rank =", rank)
print("singular_values =", singular_values)

# Plot line.
letters = X[:, 1]
sounds = y[:, 0]
x_line = np.linspace(letters.min() - 1, letters.max() + 1, 100)
y_line = b[0, 0] + b[1, 0] * x_line

plt.figure()
plt.scatter(letters, sounds)
plt.plot(x_line, y_line)
plt.xlabel("Number of letters")
plt.ylabel("Number of speech sounds")
plt.title("Least-squares line from pseudoinverse")
plt.tight_layout()
plt.show()
```

```python
# Linear regression with scikit-learn

from sklearn.linear_model import LinearRegression

# scikit-learn does not need the extra column of 1s.
# It estimates the intercept separately by default.
X = np.array([4, 8, 7, 5]).reshape(-1, 1)
y = np.array([4, 4, 5, 3])

reg = LinearRegression()
mod = reg.fit(X, y)

print("X =\n", X)
print("y =", y)
print("slope/coefficient =", mod.coef_)
print("intercept =", mod.intercept_)
print("predictions =", mod.predict(X))
print("R^2 score =", mod.score(X, y))
```
