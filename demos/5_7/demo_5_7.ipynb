```python
import numpy as np

# For a matrix, each inner list is one row.
A = np.array([[1, 2], [3, 4]])
B = np.array([[1, 3], [2, 4]])
print("A =\n", A)
print("B =\n", B)

# Column vectors are often written as matrices with one column.
a = np.array([[1], [2], [3]])
b = np.array([[4], [5]])
print("a =\n", a)
print("b =\n", b)

print("A.shape =", A.shape)
print("a.shape =", a.shape)
```

```python
A = np.array([[1, 2], [3, 4]])
A_T = A.T
print("A =\n", A)
print("A.T =\n", A_T)

a = np.array([[1], [2], [3]])
a_T = a.T
print("a =\n", a)
print("a.T =\n", a_T)
```

```python
a = np.array([[1], [2], [3]])
b = np.array([[4], [5], [6]])

# Dot product = multiply corresponding elements and add.
# 1*4 + 2*5 + 3*6 = 32
by_hand = 1 * 4 + 2 * 5 + 3 * 6

print("by hand:", by_hand)
print("np.dot(a.T, b):\n", np.dot(a.T, b))
print("a.T.dot(b):\n", a.T.dot(b))
print("a.T @ b:\n", a.T @ b)

# For 1D arrays, NumPy returns a scalar-like value instead of a 1x1 matrix.
# a_1d = np.array([1, 2, 3])
# b_1d = np.array([4, 5, 6])
# print("a_1d @ b_1d:", a_1d @ b_1d)
# print("a_1d @ b_1d:", np.dot(a_1d, b_1d))
```

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print("A =\n", A)
print("B =\n", B)
print("A @ B =\n", A @ B)

# Dimension mismatch example.
C = np.array([[1, 2, 3], [4, 5, 6]])  # shape 2 x 3
D = np.array([[1, 2], [3, 4]])        # shape 2 x 2
print("C.shape =", C.shape)
print("D.shape =", D.shape)
try:
    print(C @ D)
except ValueError as e:
    print("C @ D gives an error because the inner dimensions do not match:")
#     print(e)
```

```python
A = np.array([[1, 2], [3, 4]])
I2 = np.eye(2)
I3 = np.eye(3)

print("2x2 identity matrix:\n", I2)
print("3x3 identity matrix:\n", I3)
print("A @ I2 =\n", A @ I2)
print("I2 @ A =\n", I2 @ A)
```

```python
A = np.array([[1, 2], [3, 4]])
Ainv = np.linalg.inv(A)

print("A =\n", A)
print("A inverse =\n", Ainv)
print("A @ Ainv =\n", A @ Ainv)
print("Ainv @ A =\n", Ainv @ A)
print("Close to identity?", np.allclose(A @ Ainv, np.eye(2)))

# # Not every matrix has an inverse.
# singular = np.array([[1, 2], [2, 4]])
# non_square = np.array([[1, 2], [3, 4], [5, 6]])
# print("\nSingular matrix =\n", singular)
# print("\nNon-square matrix =\n", non_square)

# try:
#     np.linalg.inv(singular)
#     # np.linalg.inv(non_square)
# except np.linalg.LinAlgError as e:
#     print("Cannot invert this matrix:", e)
```
